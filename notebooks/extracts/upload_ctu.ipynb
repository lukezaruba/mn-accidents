```python
import geopandas as gpd
from sqlalchemy import create_engine
import geoalchemy2
import os
```

```python
# Read in SHP
gdf = gpd.read_file(r"../data/core/final_ctu_data_4326.shp")
```

```python
# Casting & Renaming
dtypes_dict = {
    "name": str,
    "class": str,
    "county_cd": int,
    "county": str
}

gdf = gdf.astype(dtypes_dict)

gdf.columns = ["gnis", "ctu_name", "class", "county_gnis", "county_code", "county", "pop", "total_road_length", "aadt_sum", "aadt_mean", "geom"]

gdf
```

```python
# Uploading to PostgreSQL
engine = create_engine(f'postgresql://{os.environ["DB_USER"]}:{os.environ["DB_PASSWORD"]}@{os.environ["DB_HOST"]}:{os.environ["DB_PORT"]}/{os.environ["DB_NAME"]}') 

gdf = gdf.set_geometry("geom")

# gdf.to_postgis("geo_ctu_accidents", engine, if_exists="append")
```
